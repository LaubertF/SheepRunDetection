```python
import pandas as pd
import matplotlib.pyplot as plt
import os
#open csv
csvpath = os.path.join('Resources', 'premiera_test2.csv')
premiera_df = pd.read_csv(csvpath)
```

```python
state = pd.read_csv('Resources/state.csv')

def add_states(plot):
    for i in range(len(state)):
        if state["State"][i] == "Run":
            color = "lime"
        else:
            color = "y"
        plot.axvspan(state['Start'][i], state['End'][i], facecolor=color, alpha=0.5)
```

```python
import matplotlib.patches as patches
from matplotlib.patches import Rectangle
def chart(data):
    start_walking = "01-09-2022 10:17:13.000" #12150
    start_running = "01-09-2022 10:17:15.000" #12150
    end_running = "01-09-2022 10:17:22.000"#12900
    end_walking2 = "01-09-2022 10:17:26.000" #12900
    #draw highlight in range
    
    x=premiera_df['datetime']
    y=data #premiera_df['gyrz'][12000:]
    plt.figure(figsize=(10,5))
    plt.plot(x,y, color='red', marker='o')
    plt.title(data.name)
    plt.xlabel('datetime')
    plt.ylabel('accx')
    #show x label every 1000
    plt.xticks(x[::800])
    plt.xticks(rotation=90)
    #add rectangle
    add_states(plt)
    plt.show()
```

```python
chart(premiera_df['accx'])
```

```python
#iterate over columns
for col in premiera_df.columns:
    if col == 'datetime':
        continue
    print(col)
    chart(premiera_df[col])
```
